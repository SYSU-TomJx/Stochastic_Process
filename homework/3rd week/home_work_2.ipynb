```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
def get_binom(n,p):
    binom_dis=stats.binom(n,p)
    sim_res = binom_dis.rvs(size=10000)
    sim_res = np.sort(sim_res)
    prob = binom_dis.cdf(sim_res)
    prob = np.unique(prob).tolist()
    return sim_res, prob, binom_dis
```

```python
def create_y_z(xn,n,p):
    
    expection = n*p
    print('n = ', n,' p = ', p)
    print('expection=',expection)
    print('average=',np.average(xn))
    print('========================')
    s_i = []
    y = [(x-expection)/n for i,x in enumerate(xn)]

    z = [(x-expection)/np.sqrt(n) for i,x in enumerate(xn)]
    y = np.unique(y).tolist()
    z = np.unique(z).tolist()
    return y, z
```

```python
xn_1,fx_1, binom_dis_1 = get_binom(4, 0.25)
xn_2,fx_2, binom_dis_2 = get_binom(20, 0.25)
xn_3,fx_3, binom_dis_3 = get_binom(50, 0.25)

yn_1,zn_1 = create_y_z(xn_1,4,0.25)
yn_2,zn_2 = create_y_z(xn_2,20,0.25)
yn_3,zn_3 = create_y_z(xn_3,50,0.25)
```

```python
plt.step(yn_1+[0.8],[0]+fx_1[:len(fx_1)-1]+[fx_1[-1]],label='n=4',marker='o',linestyle="-")
plt.step(yn_2+[0.8], [0]+fx_2[:len(fx_2)-1]+[fx_2[-1]], label='n=20',marker='s',linestyle="-.")
plt.step(yn_3+[0.8], [0]+fx_3[:len(fx_3)-1]+[fx_3[-1]], label='n=50',marker='^',linestyle=":")
plt.legend() 
plt.title("The CDF of $Y_n$")
plt.show()
```
