```python
x_un = []
s_un = []
for i in range(5):
    x_i_un = np.random.uniform(low=0.0, high=1.0, size=(n))
    x_i_un = np.around(x_i_un,decimals=3)
    s_i_un = [np.sum(x_i_un[0:i]) for i in range(1,n+1)]
    s_i_un = np.around(s_i_un, decimals=3)
    print('x_'+str(i+1)+' = ', x_i_un,', s_'+str(i+1)+'= ', s_i_un)
    plt.step(s_i_un, n_i[1:], label=str(i+1), marker=MARKS[i])
plt.legend()
plt.title('Renewal Processes')
plt.xlabel('$S_n$')
plt.show()
```

#### 3. 泊松过程$\lambda$=1。

```python
n = 10
for i in range(5):
    x_i_po = np.random.poisson(lam=1.0, size=(n))
    s_i_po = [np.sum(x_i_po[0:i]) for i in range(1,n+1)]
    print('x_'+str(i+1)+' = ', x_i_po,', s_'+str(i+1)+'= ', s_i_po)
    plt.step(s_i_po, n_i[1:], label=str(i+1), marker=MARKS[i])
plt.legend()
plt.xlabel('$S_n$')
plt.title('Poission Processes')
plt.show()
```
